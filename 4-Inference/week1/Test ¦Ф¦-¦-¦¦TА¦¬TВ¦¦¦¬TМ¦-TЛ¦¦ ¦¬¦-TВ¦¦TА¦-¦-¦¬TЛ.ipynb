```python
sample_test=[1]*s[0]+[0]*(n[0]-s[0])
sample_control=[1]*s[1]+[0]*(n[1]-s[1])
```

```python
def proportions_confint_diff_ind(sample1, sample2, alpha = 0.05):    
    z = scipy.stats.norm.ppf(1 - alpha / 2.)   
    p1 = float(sum(sample1)) / len(sample1)
    p2 = float(sum(sample2)) / len(sample2)
    
    left_boundary = (p1 - p2) - z * np.sqrt(p1 * (1 - p1)/ len(sample1) + p2 * (1 - p2)/ len(sample2))
    right_boundary = (p1 - p2) + z * np.sqrt(p1 * (1 - p1)/ len(sample1) + p2 * (1 - p2)/ len(sample2))
    
    return (left_boundary, right_boundary)
```

```python
btm,top=proportions_confint_diff_ind(sample_control,sample_test)
print (btm,top)
round(top,4)
```

```python
def odds(sample):
    p=np.sum(sample)/len(sample)
    return p/(1-p)
def odds_rel(sample_control,sample_test):
    odds_control=odds(sample_control)
    odds_test=odds(sample_test)
    return odds_control/odds_test
round(odds_rel(sample_control,sample_test),4)
```

```python
def get_bootstrap_samples(data, n_samples):
    indices = np.random.randint(0, len(data), (n_samples, len(data)))
    samples = data[indices]
    return samples
def stat_intervals(stat, alpha):
    boundaries = np.percentile(stat, [100 * alpha / 2., 100 * (1 - alpha / 2.)])
    return boundaries
sample_control=np.array(sample_control)
sample_test=np.array(sample_test)
```

```python
np.random.seed(0)
num_boot=1000
odds_list=[]
for bs_control,bs_test in zip(get_bootstrap_samples(sample_control,num_boot),get_bootstrap_samples(sample_test,num_boot)):
    odds_list.append(odds_rel(bs_control,bs_test))
odds_list
```

```python
ci_b,ci_t=stat_intervals(odds_list, 0.05)
print( "95% confidence interval",  ci_b,ci_t)
```

```python
round(ci_b,4)
```
