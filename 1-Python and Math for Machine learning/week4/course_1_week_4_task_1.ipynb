# Центральная предельная теорема своими руками

```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%pylab inline
```

## Определение функции, отрисовка графиков

```python
# Выбираем гамма распределение
a = 9
b = 0.5
rv = sts.gamma(a,scale=b)
# Генерируем выборку объёма 1000
N=1000
sample=rv.rvs(N)
```

```python
# Вычисляем плотность распределения
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 100)
pdf=rv.pdf(x)
```

```python
# Строим гистограмму выборки
plt.hist(sample,normed=True,label='gamma hist')
# Строим гистограмму выборки
plt.plot(x, pdf,'r-', lw=5, alpha=0.7, label='gamma pdf')
plt.legend(loc='best', frameon=False)
plt.xlabel('sample')
plt.ylabel('f(x)')
plt.title("Гистограмма и плотность Гамма распредленения ")
plt.show()
```

## Оценка распределения выборочного среднего при разных объёмах выборок

```python
# Посчитаем параметры нормального распределения

# Сначала вычислим теоретические значения мат.ожидания и дисперсии для Гамма распределения
# Для гамма распределения G(a,b) - матожидание = a*b, дисперсия = a*b*b

# Дисперсия 
variance = a*b*b
print ("Дисперсия=",variance)
# Математическое ожидание
expect= a*b 
print("Матожидание=",expect)
```

```python
# Дисперсия для нормального распредление
norm_variance=math.sqrt(variance)
norm_rv = sts.norm(expect,norm_variance)
norm_x = np.linspace(norm_rv.ppf(0.01), norm_rv.ppf(0.99), 100)
norm_pdf = norm_rv.pdf(norm_x)
plt.plot(x, norm_pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
f, axarr = plt.subplots(nrows=3, ncols=2)
f.set_figheight(15)
f.set_figwidth(15)

# Количество выборок
qty=1000
# Будем формировать выборки разных объемов n
for ax_n,n in enumerate([5,10,50,250,500,1250]):

    means=list()
    # Фомировать 1000 выборок
    for i in range(qty):
        # Формируем выборку объема n
        sample=rv.rvs(n)
        # Считаем выборочное среднее для выборки
        means.append(np.mean(sample))
    # Дисперсия для апроксимирующего нормального распредление
    norm_variance=math.sqrt(variance/n)
    # Вычислим плотность на том же множестве
    norm_rv = sts.norm(expect,norm_variance)
    norm_x = np.linspace(norm_rv.ppf(0.01), norm_rv.ppf(0.99), 100)
    norm_pdf = norm_rv.pdf(norm_x)
    
    #Строим гистограму распределений значений выборочного срденего для очередного n и функцию лотности для нормального распредления
    ax_row=ax_n%3
    ax_col=int(ax_n/3)
    axarr[ax_row,ax_col].plot(norm_x,norm_pdf,label='normal pdf')
    axarr[ax_row,ax_col].hist(means,bins=round(qty/50),normed=True,label='gamma hist')
    axarr[ax_row,ax_col].set_title("n="+str(n))
    axarr[ax_row,ax_col].set_xlabel('sample')
    axarr[ax_row,ax_col].set_ylabel('f(x)')
    axarr[ax_row,ax_col].legend(loc='best', frameon=False)
    
plt.show()
```

### Вывод 
Действительно, распределение выборочных средних представляет собой нормальное распределение с указанными параметрами, в чем можно убедиться по данным графикам. Мы построили гистограмму данных гамма-распределения и график плотности вероятности нормального распределения при разных объемах выборки. Можем видеть, что плотность вероятности достаточно точно оценивается, причем с ростом объема выборки точность аппроксимации улучшается.
