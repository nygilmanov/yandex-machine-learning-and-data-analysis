```python
#; По входному вектору X формируем матрицу коэффециентов A
def genA(x):
    step_slau=len(x)
    l=list()
    # Цикл по строчкам матрицы
    for i in x:
        # Цикл по столбцам
        for j in range(step_slau):
           
            l.append(i**j)
    n=np.array(l)
    n=n.reshape(step_slau,step_slau)
    return n
# По входному вектору X формируем вектор значений фуникции b
def genB(x):
    l=list()
    for i in x:
        l.append(func(i))
    n=np.array(l)
    return n
# По входному вектору w формируем значение функции многочлена 
def genF(w,x):
    f=0
    for i in range(len(w)):
        f=f+w[i]*x**i
    return f
```
