```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.tsa.stattools as stat
import sklearn.linear_model as sk
import itertools
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

## Исследуем район 1282

```python
# Загружаем готовую таблицу за январь-май 2016 по самому активному району - 1281
data = pd.read_csv('region_1282_date.csv', sep=';', index_col='Unnamed: 0')
data.shape
```

```python
# Делаем первую визуализацию
plt.figure(figsize=(20,5))
data.rides.plot()
plt.title(u'Поездки из 1282 по часам за янв-май 2016')
plt.xlabel(u'Время')
```

```python
# Визуально видно, что дисперсия в течение месяца меняется
# Проводим тест Дики Фулера на стационарность
print('Level of pvalue:', stat.adfuller(data.rides)[1])
```

Формально, при уровне значимости 0.05, мы можем отвергнуть гепотизу о нестационарности. Но визуально это не подтверждается. Нужно провести работу по исключению недельной сезонности.

## Создаем новые целевые признаки и работаем с ними

Делаем ряды синуса и косинуса для дальнейшего использования их как признаки в регрессионной модели.

```python
T = 3648 # Параметр равен количеству часов во временном ряду
Ks1, Ks2, Ks3, Kc1, Kc2 = [1., 7., 14., 1., 7.]

s1 = np.sin(np.array(range(1,T+1)) * 2.*np.pi*Ks1/168.)
s2 = np.sin(np.array(range(1,T+1)) * 2.*np.pi*Ks2/168.)
s3 = np.sin(np.array(range(1,T+1)) * 2.*np.pi*Ks3/168.)

c1 = np.cos(np.array(range(1,T+1)) * 2.*np.pi*Kc1/168.)
c2 = np.cos(np.array(range(1,T+1)) * 2.*np.pi*Kc2/168.)
```

```python
s1.shape
```

Делаем линейную регрессионную модель и оцениваем ее остатки.

```python
# Объединяем признаки в единую матрицу X
x1 = s1.reshape(T,1)
x2 = s2.reshape(T,1)
x3 = s3.reshape(T,1)
x4 = c1.reshape(T,1)
x5 = c2.reshape(T,1)
X = np.hstack((x1,x2,x3,x4,x5))
```

```python
# Обучаем линейную модель и находим остатки
y = list(data.rides)
est_LR = sk.LinearRegression()
est_LR.fit(X, y)
y_pred = est_LR.predict(X)
residuals = y - y_pred
```

```python
# Выводим график остатков
plt.figure(figsize=(20,5))
plt.plot(residuals)
plt.title(u'Остатки линейной регрессии на новые признаки')
plt.xlabel(u'Время')
```

```python
print('Pvalue of DF test on the residuals:', stat.adfuller(residuals)[1])
```

#### Примечание:
Использование 5 дополнительных признаков в виде колебаний sin и cos полностью убрали недельную сезонность. Остатки визуально стационарны, что подтверждается критерием Дики Фулера, который упал на один порядок.

## STL декомпозиция и дифференцирование

```python
# Добавляем остатки в изначальные данные
# Делаем дифференцирование на 1 неделю (168 часов)
data['res'] = residuals
data['res_d'] = data.res - data.res.shift(168)
```

Смотрим на результаты дифференцирования.

```python
# Для построения STL декомпозиции необходимо сменить формат индекса
data.reset_index(inplace=True)
data['index'] = pd.to_datetime(data['index'])
data = data.set_index('index')
```

```python
# Делаем STL декомпозицию
stl_res = sm.tsa.seasonal_decompose(data.res_d[168:])

fig, axarr = plt.subplots(3, sharex=True, figsize=(20,15))
axarr[0].plot(stl_res.trend)
axarr[0].set_title(u'Тренд')
axarr[1].plot(stl_res.seasonal)
axarr[1].set_title(u'Сезонные колебания')
axarr[2].plot(stl_res.resid)
axarr[2].set_title(u'Остатки')
```

#### Примечание:
Дифференцирование на 1 неделю дополнительно стабилизирует остатки, но непринципильно. Однозначный тренд, при этом, отсутствует.

## Подбор параметров модели ARIMA

```python
# Строим графики ACF и PACF для подбора стартового уровня параметров
fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data.res_d[168:], lags=168, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data.res_d[168:], lags=168, ax=ax2)
```

```python
# Возьмем стартовыми следующие параметры
qs = range(0,10)
Qs = range(0,4)
ps = range(0,3)
Ps = range(0,2)
d = 0
D = 0
s = 24
```

#### Примечание:
Как видно, было решено отказаться от диференцирования, поскольку: 1) ряд остатков и так достаточно стационарен; 2) наличие диференцирования замедляет работу модели; 3) подбор параметров без дифференцирования дает замечательные результаты.

## Делаем модель SARIMA

```python
parameters = itertools.product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")

for param in parameters_list:
    # Подбор параметров будем делать на остатках, что бы ускорить работу модели
    try:
        model=sm.tsa.statespace.SARIMAX(data.res, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], s),
                                        enforce_stationarity=False,
                                        enforce_invertibility=False).fit(disp=-1)

    except ValueError:
        print('wrong parameters:', param)
        continue
    print('good parameters:', param)
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
# Подтверждаем, что модель с минимальным AIC
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
# Смотрим на параметры модели
print(best_model.summary())
```

```python
# Визуальная оценка качества модели
data['model'] = best_model.fittedvalues
data['filled'] = data['model'].fillna(0)
data['LR'] = y_pred
data['total'] = data['filled'] + data['LR']

plt.figure(figsize=(20,5))
data.rides.plot()
data.total.plot(color='r')
plt.title(u'Сравнение фактического и предсказанного количества поездок')
plt.ylabel(u'Позедки')
plt.show()
```

#### Примечание:
Модель хорошо предсказывает все исходный временной ряд. Возможно это связано с тем, что колебание количества поездок в самом популярном районе слабо зависит от прочих факторов.
