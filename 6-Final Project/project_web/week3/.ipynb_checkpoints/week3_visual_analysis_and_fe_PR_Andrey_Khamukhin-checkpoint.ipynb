**Создайте на основе функций *prepare_train_set* и *prepare_sparse_train_set_window*  новую – *prepare_train_set_with_fe*, (от "feature engineering"), создайте следующие признаки:**
- session_timespan – продолжительность сессии (разница между максимальным и минимальным временем посещения сайтов в сессии, в секундах)
- #unique_sites – число уникальных сайтов в сессии 
- start_hour – час начала сессии (то есть час в записи минимального timestamp среди десяти)
- day_of_week – день недели (то есть день недели в записи минимального timestamp среди десяти)

Функция должна возвращать новый DataFrame (как возвращала функция *prepare_train_set*), только признаков должно быть на 4 больше. Порядок, в котором добавляются признаки: *site1*, ... *site10*, *session_timespan*, *#unique_sites*, *start_hour*, *day_of_week* и *user_id* (это видно и чуть ниже по тому, как функция вызывается).

```python
def prepare_train_set_with_fe(path_to_csv_files, site_freq_path, feature_names,
                                    session_length=10, window_size=10):
    ''' ВАШ КОД ЗДЕСЬ '''
    from time import time
    from math import ceil
    init_time = time()
    
    with open(site_freq_path, 'rb') as site_freq_pkl:
        site_ids_and_freqs = pickle.load(site_freq_pkl)
        
    site_ids_and_freqs[0] = (0, 0)    
    users, all_sessions = [], []
    timespans = []
    num_unique_sites = []
    start_hours, days_of_week = [], [] 
    
    for file_id, file_name in enumerate(sorted(glob(os.path.join(path_to_csv_files, '*')))):
        data = pd.read_csv(file_name)
        user_id = int(file_name.strip('.csv')[-4:])
        
        times = data['timestamp']
        
        site_ids = data['site'].apply(lambda site: site_ids_and_freqs[site][0])
    
        for session_start_index in range(0, len(site_ids.values), window_size):
            
            new_session = site_ids.values[session_start_index: 
                                               session_start_index + session_length].tolist()
            session_times = times[session_start_index:
                                 session_start_index + session_length]
            start_time = pd.to_datetime(session_times.iloc[0])
            
            timespans.append((pd.to_datetime(session_times.iloc[-1]) - 
                             start_time).total_seconds())
            num_unique_sites.append(int(np.unique(new_session).shape[0]))
            start_hours.append(start_time.hour)
            days_of_week.append(start_time.dayofweek)
            
            if len(new_session) < session_length:
                new_session_length = len(new_session)
                for _ in range(session_length - new_session_length):
                    new_session.append(0)
                    
            all_sessions.append(new_session)
            users.append(user_id)

    num_sess = len(all_sessions)
    X = np.array(all_sessions).astype('int')
    timespans = np.array(timespans).reshape([num_sess, 1]).astype('int')
    num_unique_sites = np.array(num_unique_sites).reshape([num_sess, 1]).astype('int')
    start_hours = np.array(start_hours).reshape([num_sess, 1]).astype('int')
    days_of_week= np.array(days_of_week).reshape([num_sess, 1]).astype('int')
    y = np.array(users).reshape([num_sess, 1])
    
    train_df = pd.DataFrame(np.hstack([X, timespans, num_unique_sites,
                                       start_hours, days_of_week, y]), 
                            columns=feature_names).drop_duplicates()  
    return train_df
```

**Напишите функцию для создания новых признаков и примените ее к исходным данным – каталогам с 10 и 150 файлами. Сделайте это только для набора данных, полученного с параметрами *session_length=10* и *window_size=10*. Сериализуйте полученные матрицы с помощью pickle. Функция может возвращать как только новые признаки, так и старые с новыми. При этом сигнатура функции может быть другой – тут уже свобода выбора.**

```python
def feature_engineering(path_to_csv_files, site_freq_path, features,
                                    session_length=10, window_size=10):
    with open(site_freq_path, 'rb') as site_freq_pkl:
        site_ids_and_freqs = pickle.load(site_freq_pkl)
        
    site_ids_and_freqs[0] = (0, 0)    
    users, all_sessions = [], []
    timespans = []
    timespans_reciprocal = []
    num_unique_sites = []
    start_hours, days_of_week = [], []
    top30_site_ids = [x[1][0] for x in sorted(site_freq_10users.items(), key=lambda x: x[1][1],reverse=True)[:30]]
    top_sites_freqs = []
    
    for file_id, file_name in enumerate(sorted(glob(os.path.join(path_to_csv_files, '*')))):
        data = pd.read_csv(file_name)
        user_id = int(file_name.strip('.csv')[-4:])
        
        times = data['timestamp']
        
        site_ids = data['site'].apply(lambda site: site_ids_and_freqs[site][0])
    
        for session_start_index in range(0, len(site_ids.values), window_size):
            
            new_session = site_ids.values[session_start_index: 
                                               session_start_index + session_length].tolist()
            session_times = times[session_start_index:
                                 session_start_index + session_length]
            start_time = pd.to_datetime(session_times.iloc[0])
            
            
            timespans.append((pd.to_datetime(session_times.iloc[-1]) - 
                             start_time).total_seconds())
            timespan_reciprocal = ((pd.to_datetime(session_times).shift(-1)-\
                                    pd.to_datetime(session_times))[:-1]).apply(
                                        lambda x: int(x.total_seconds()) if x.total_seconds()<5*3600 else 0).values.tolist()
            num_unique_sites.append(int(np.unique(new_session).shape[0]))
            start_hours.append(start_time.hour)
            days_of_week.append(start_time.dayofweek)
            
            if len(new_session) < session_length:
                new_session_length = len(new_session)
                for _ in range(session_length - new_session_length):
                    new_session.append(0)
                    timespan_reciprocal.append(0)
            
            sites_count = np.zeros(len(top30_site_ids))
            sites_avg_time = np.zeros(len(top30_site_ids),dtype=float)
            last_site_id = new_session[0]
            last_site_time = pd.to_datetime(session_times.iloc[0])
            for i,site_id,site_time in zip(range(1,session_length),new_session[1:],session_times[1:]):
                if last_site_id != site_id:
                    if last_site_id in top30_site_ids:
                        delta_time = (pd.to_datetime(session_times.iloc[i])-last_site_time).total_seconds()
                        site_index = top30_site_ids.index(last_site_id)
                        sites_count[site_index] += 1
                        sites_avg_time[site_index] = (sites_avg_time[site_index]*(sites_count[site_index]-1)+delta_time)/sites_count[site_index]
                    last_site_time = pd.to_datetime(session_times.iloc[i])
                    last_site_id = site_id
            
            top_sites_freqs.append(sites_avg_time.tolist())
            timespans_reciprocal.append(timespan_reciprocal)
            all_sessions.append(new_session)
            users.append(user_id)

    num_sess = len(all_sessions)
    X = np.array(all_sessions).astype('int')
    timespans = np.array(timespans).reshape([num_sess, 1]).astype('int')
    num_unique_sites = np.array(num_unique_sites).reshape([num_sess, 1]).astype('int')
    start_hours = np.array(start_hours).reshape([num_sess, 1]).astype('int')
    days_of_week= np.array(days_of_week).reshape([num_sess, 1]).astype('int')
    y = np.array(users).reshape([num_sess, 1])
    
    train_df = pd.DataFrame(np.hstack([X, timespans, num_unique_sites,
                                       start_hours, days_of_week, y, top_sites_freqs, timespans_reciprocal]), 
                            columns=features).drop_duplicates()  
    return train_df
```

```python
feature_names = ['site' + str(i) for i in range(1,11)] + \
                ['session_timespan', '#unique_sites', 'start_hour', 'day_of_week', 'user_id']
new_feature_names = ['top' + str(i) for i in range(1,31)] + \
                    ['timespan' + str(i) for i in range(1,10)]
```
