```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%pylab inline
```

```python
#генерирование гамма-распределения с параметрами 1, 2
gamma = np.random.gamma(1, 2, 1000)
#гистограмма данных
count, bins, ignored = plt.hist(gamma, 50, normed = True)
#график плотности вероятности гамма-распределения с данными параметрами
plt.plot(bins, sts.gamma.pdf(bins, 1, loc = 0, scale = 2), label = 'gamma pdf', linewidth = 2, color = 'r')
plt.legend(loc = 'best', frameon = False)

plt.show()
```

```python
x = []
for i in range(1000):
    gamma = np.random.gamma(1, 2, 5)
    x.append(sum(gamma) / len(gamma))
#гистограмма данных
count, bins, ignored = plt.hist(x, 50, normed = True)
#математическое ожидание = 1*2; дисперсия = 1*(2)^2
#график плотности вероятности нормального распределения с данными МО и дисперсией
plt.plot(bins, sts.norm.pdf(bins, loc = 2, scale = np.sqrt(4/5)), label = 'normal pdf', linewidth = 2, color = 'r')
plt.legend(loc = 'best', frameon = False)
plt.show()
```

```python
x = []
for i in range(1000):
    gamma = np.random.gamma(1, 2, 10)
    x.append(sum(gamma) / len(gamma))
#гистограмма данных
count, bins, ignored = plt.hist(x, 50, normed = True)
#математическое ожидание = 1*2; дисперсия = 1*(2)^2
#график плотности вероятности нормального распределения с данными МО и дисперсией
plt.plot(bins, sts.norm.pdf(bins, loc = 2, scale = np.sqrt(4/10)), label = 'normal pdf', linewidth = 2, color = 'r')
plt.legend(loc = 'best', frameon = False)
plt.show()
```

```python
x = []
for i in range(1000):
    gamma = np.random.gamma(1, 2, 50)
    x.append(sum(gamma) / len(gamma))
#гистограмма данных
count, bins, ignored = plt.hist(x, 50, normed = True)
#математическое ожидание = 1*2; дисперсия = 1*(2)^2
#график плотности вероятности нормального распределения с данными МО и дисперсией
plt.plot(bins, sts.norm.pdf(bins, loc = 2, scale = np.sqrt(4/50)), label = 'normal pdf', linewidth = 2, color = 'r')
plt.legend(loc = 'best', frameon = False)
plt.show()
```

### Выкладки с вычислениями параметров нормального распределения

Параметры гамма распределения $k = 1; \theta = 2$

Как мы знаем, распределение выборочных средних представляет собой нормальное распределение с параметрами $\mu = k\theta, \sigma^2 = \frac{k\theta^2}{n}$, где n - объем выборки 

### Вывод

Действительно, распределение выборочных средних представляет собой нормальное распределение с указанными параметрами, в чем можно убедиться по данным графикам. Мы построили гистограмму данных гамма-распределения и график плотности вероятности нормального распределения при разных объемах выборки. Можем видеть, что плотность вероятности достаточно точно оценивается, причем с ростом объема выборки точность аппроксимации улучшается.
