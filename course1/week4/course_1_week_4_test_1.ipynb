# Тренировочный тест
## ЦПТ и доверительные интервалы

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 1197.69
sigma_kv=116182.2
sigma = math.sqrt(116182.2)
start_year=1885
end_year=2015
years=end_year-start_year+1
# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)

# сгенерируем years значений
norm_rv.rvs(size=years)
```

```python
x = np.linspace(mu-3*sigma,mu+3*sigma,years)
cdf = norm_rv.cdf(x) # функция может принимать и вектор (x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
x = np.linspace(mu-3*sigma,mu+3*sigma,years)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

## Задача 1 - Верхняя граница интервала

```python
print ("Task 1=",round(mu+3*sigma,2))
```

## Задача 2 - значение среднеквадратического отклонения нормального распределения

```python
d=sigma_kv/years
d=math.sqrt(d)
print("Task2=",round(d,2))
```

## Задача 3 - Верхний доверительный предел

```python
alpha = 1 - 0.997
z = abs(sts.norm.ppf(alpha/2))
print (z)
```

```python
SE = sigma/math.sqrt(years)
```

```python
ci_low = mu - z * SE # ci - сокращение от confidence interval
ci_hi = mu + z * SE
(ci_low, ci_hi)
```

```python
print ("Task3=",round(ci_hi,2))
```
