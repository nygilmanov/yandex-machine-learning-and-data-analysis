Необходимо проанализировать данные, подобрать для них оптимальную прогнозирующую модель в классе ARIMA и построить прогноз на каждый месяц на два года вперёд от конца данных.

#### Придерживайтесь стандартного алгоритма построения прогноза:

1) Визуальный анализ ряда

2) Стабилизация дисперсии (при необходимости)

3) Выбор порядка дифференцирования

4) Выбор начальных приближений для p,q,P,Q

5) Обучение и сравнение моделей-кандидатов, выбор победителя

6) Анализ остатков построенной модели, при необходимости — её модификация: если остатки получаются смещёнными, прогноз нужно скорректировать на константу; если остатки нестационарны или автокоррелированы, можно попробовать расширить область, в которой подбираются значения p,q,P,Q.

7) Прогнозирование с помощью финальной модели.

## Импорт данных и первая визуализация

```python
data = pd.read_csv('WAG_C_M.csv', sep=';', index_col='month')
print data.shape
print data.columns[0]
```

```python
plt.figure(figsize(10,5))
data.WAG_C_M.plot()
plt.ylabel(u'заработная плата')
plt.title(u'Исходный ряд')
```

#### Примечание:
На графике видно, что временной ряд имеет тренд, сезонность и растущую дисперсию. Ряд не является стационарным.

## Стабилизация дисперсии

Сделаем преобразование Бокса-Кокса для стабилизации дисперсии.

```python
data['wage_bx'], par_bx = stats.boxcox(data.WAG_C_M)
print par_bx
```

```python
plt.figure(figsize(10,5))
data.wage_bx.plot()
plt.title(u'Ряд после применения Бокса-Кокса')
```

#### Примечание:
Размах дисперсии получилось сжать, но дисперсия по прежнему отличается в начале и в конце ряда.

## Исключение тренда через дифференцирование

Некоторый перебор показал, что лучшим образом тренд исключается дифференцированием первого порядка.

```python
data['wage_bx_d1'] = data.wage_bx - data.wage_bx.shift(1)
```

Сразу проведем проверку на стационарность, применив критерий Дики-Фуллера.

```python
to_df1 = list(data.wage_bx_d1.dropna())
p1 = sm.tsa.stattools.adfuller(to_df1)[1]
print 'Значимость критерия Дики-Фуллера: p=%f' % p1
```

На уровне значимости 0.05 гипотеза о нестационарности отвергается. Теперь проверим полученный ряд визуально.

```python
plt.figure(figsize(10,5))
data.wage_bx_d1.plot()
plt.title(u'График первых разностей')
```

#### Примечание:
Как видно из графика, тренд полностью исключен (визуально) и разница в дисперсии сохранилась, но уменьшилась (что мы выяснили до этого). Несмотря на преобразования, ряд по прежнему нестационарен (визуально).

В связи в этим проводится повторное дифференцирование. Подбор показал, что порядок дифференцирования равный 12 дает лучший результат.

```python
data['wage_bx_d1_d12'] = data.wage_bx_d1 - data.wage_bx_d1.shift(12)
```

```python
data.wage_bx_d1_d12.plot()
```

```python
to_df2 = list(data.wage_bx_d1_d12.dropna())
p2 = sm.tsa.stattools.adfuller(to_df2)[1]
print 'Значимость критерия Дики-Фуллера: p=', p2
```

#### Примечание:
После двойного дифференцирования ряд стал визуально наиболее стационарным, что подтверждается также критерием Дики-Фуллера (гипотеза о нестационарности уверенно отвергается).

## Выбор начальных приближений

Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(data.wage_bx_d1_d12[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(data.wage_bx_d1_d12[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

#### Примечание
Подбор параметров дает следующие результаты:
Q=0, q=1, P=0, p=1

Параметр P может быть равен 0, но поскольку коррелограмма показывает, что его первое значение находится на границе зоны значимости, я его включил для подбора, на случай, если это улучшит результат. Если нет, то по критерию AIC его отбракуем.

## Обучение

```python
d  = 1
D  = 1
ps = [0,1]
qs = [0,1]
Ps = [0,1]
Qs = 0
```

```python
# Исключаем Qs, т.к. его значение фиксировано на уровне 0.
parameters = itertools.product(ps, qs, Ps)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.wage_bx, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, Qs, 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

#### Примечание:
Лучшая модель SARIMAX(p=1, q=0)x(P=1, Q=0) с обычным (d=1) и сезонным (d=12) дифференцированием.

## Проверка остатков

```python
plt.figure(figsize(10,10))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Остатки')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print('Критерий Стьюдента: p=%f' % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print 'Критерий Дики-Фуллера: p=', sm.tsa.stattools.adfuller(best_model.resid[13:])[1]
```

#### Примечание:
Остатки хорошие, стационарные, без автокорреляции.

```python
def invboxcox(y,par_bx):
   if par_bx == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(par_bx*y+1)/par_bx))
```

```python
data['model'] = invboxcox(best_model.fittedvalues, par_bx)
data['filled'] = data['model'].fillna(0) 
plt.figure(figsize(10,5))
data.WAG_C_M.plot()
data.filled.plot(color='r')
plt.ylabel('Wage')
pylab.show()
```

#### Примечание:
Ряд модели хорошо повторяет характер изначального ряда.
