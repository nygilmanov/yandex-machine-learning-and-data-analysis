```python
import math
# Выберем экспоненциальную функция распределения с lambda = 0.5
lam = 0.5
from numpy.random import exponential
# Для получния выборок будем пользоваться numpy.random.exponential
# При этом необходимо передать параметр beta = 1 / lambda
beta = 1.0 / lam
# Матожидание и дисперсия экспоненциального распределения
EX = 1 / lam
DX = 1 / (lam * lam)
```

```python
# Экспоненциальная функция распределения (совесем не похожа на нормальное)
def f(x):
    if x < 0:
        return 0
    else:
        return lam * math.exp(-lam * x)
```

```python
# Выборка из экспоненциальной функции распределения
def sample(N):
    return [exponential(beta) for x in xrange(N)]
```

```python
import matplotlib.pyplot as plt, numpy as np
%matplotlib inline 

plt.hist(sample(1000), 30, normed=True)
xpoints = np.linspace(0, 10, 1000)
plt.plot(xpoints, [f(x) for x in xpoints], linewidth=3)
plt.ylabel('probability density')
plt.xlabel('value')
plt.show()

# Для отладки, изображение гистограммы выборки с экспоненциальным расперделением
# и теоретическое значение плотности вероятности
# Видно что похожи, значит мы не напутали с lambda и beta
```

```python
# Вычисление выорочного среденего
def sampleMean(N):
    return sum(sample(N)) / N
```

```python
# Вычисление среднего из выорочного среденего
def sampleOfSampleMeans(N, volume):
    return [sampleMean(volume) for x in xrange(N)]    
```

```python
# Сравнение распределения выборочных средних на выборке объема 5
# с нормальным распределением c соответсвующим мат. ожиданием и дисперсией
from scipy.stats import norm
plt.hist(sampleOfSampleMeans(1000, 5), 30, normed=True)
plt.plot(xpoints, [norm.pdf(x, EX, math.sqrt(DX / 5)) for x in xpoints], linewidth=3)
plt.xlabel('sample mean')
plt.ylabel('probability density')
plt.show()
```

```python
# Сравнение распределения выборочных средних на выборке объема 10
# с нормальным распределением c соответсвующим мат. ожиданием и дисперсией

plt.hist(sampleOfSampleMeans(1000, 10), 30, normed=True)
plt.plot(xpoints, [norm.pdf(x, EX, math.sqrt(DX / 10)) for x in xpoints], linewidth=3)
plt.xlabel('sample mean')
plt.ylabel('probability density')
plt.show()
```

```python
# Сравнение распределения выборочных средних на выборке объема 50
# с нормальным распределением c соответсвующим мат. ожиданием и дисперсией
plt.hist(sampleOfSampleMeans(1000, 50), 30, normed=True)
plt.plot(xpoints, [norm.pdf(x, EX, math.sqrt(DX / 50)) for x in xpoints], linewidth=3)
plt.xlabel('sample mean')
plt.ylabel('probability density')
plt.show()
```

```python
# Эксперементы показывают, что при увеличении объема выборки, распределение выборочных 
# средниx приближаются к нормальному распределению с соотвествующей дисперсией и мат. ожиданием
```
